```python
# JUPYTER NOTEBOOK CELL: Milkshake vs Cake Shop (UK)  menus, demographics, and profit simulation

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------------
# 1) MENUS (synthetic but realistic-leaning)
# ----------------------------

def build_menus():
    milkshake_menu = pd.DataFrame([
        {"item":"Classic Vanilla Shake (12oz)", "category":"classic", "price_gbp":4.95, "cogs_gbp":1.25, "prep_min":3},
        {"item":"Classic Chocolate Shake (12oz)", "category":"classic", "price_gbp":4.95, "cogs_gbp":1.30, "prep_min":3},
        {"item":"Strawberry Shake (12oz)", "category":"classic", "price_gbp":5.25, "cogs_gbp":1.35, "prep_min":3},
        {"item":"Salted Caramel Shake (16oz)", "category":"premium", "price_gbp":6.25, "cogs_gbp":1.85, "prep_min":4},
        {"item":"Oreo / Cookies Shake (16oz)", "category":"premium", "price_gbp":6.75, "cogs_gbp":2.10, "prep_min":4},
        {"item":"Protein Shake (16oz)", "category":"functional", "price_gbp":7.25, "cogs_gbp":2.55, "prep_min":4},
        {"item":"Vegan Shake (oat-based, 16oz)", "category":"dietary", "price_gbp":6.95, "cogs_gbp":2.35, "prep_min":4},
        {"item":"Add-on: Whipped cream", "category":"addon", "price_gbp":0.75, "cogs_gbp":0.18, "prep_min":0.5},
        {"item":"Add-on: Extra syrup", "category":"addon", "price_gbp":0.60, "cogs_gbp":0.12, "prep_min":0.2},
        {"item":"Add-on: Extra topping (sprinkles/cookie)", "category":"addon", "price_gbp":0.85, "cogs_gbp":0.25, "prep_min":0.3},
    ])

    cake_menu = pd.DataFrame([
        {"item":"Victoria Sponge (slice)", "category":"slice", "price_gbp":3.95, "cogs_gbp":1.10, "prep_min":1.2},
        {"item":"Chocolate Fudge Cake (slice)", "category":"slice", "price_gbp":4.25, "cogs_gbp":1.25, "prep_min":1.2},
        {"item":"Carrot Cake (slice)", "category":"slice", "price_gbp":4.35, "cogs_gbp":1.30, "prep_min":1.2},
        {"item":"Cheesecake (slice)", "category":"slice", "price_gbp":4.65, "cogs_gbp":1.55, "prep_min":1.2},
        {"item":"Brownie", "category":"traybake", "price_gbp":3.25, "cogs_gbp":0.85, "prep_min":1.0},
        {"item":"Cookie", "category":"snack", "price_gbp":2.25, "cogs_gbp":0.55, "prep_min":0.8},
        {"item":"Cupcake", "category":"snack", "price_gbp":2.95, "cogs_gbp":0.70, "prep_min":1.0},
        {"item":"Celebration Cake (whole, 10 slices)", "category":"whole_cake", "price_gbp":34.00, "cogs_gbp":11.50, "prep_min":6.0},
        {"item":"Add-on: Ice cream scoop", "category":"addon", "price_gbp":1.50, "cogs_gbp":0.40, "prep_min":0.3},
        {"item":"Add-on: Sauce drizzle", "category":"addon", "price_gbp":0.70, "cogs_gbp":0.15, "prep_min":0.2},
    ])

    return milkshake_menu, cake_menu


milkshake_menu, cake_menu = build_menus()

display(milkshake_menu)
display(cake_menu)

# ----------------------------
# 2) CUSTOMER DEMOGRAPHICS + BUYING LOGIC
# ----------------------------
# We'll model the UK customer mix as segments with:
# - share of footfall
# - probability of purchase
# - typical basket size + add-on behaviour
# - preferred categories

segments = pd.DataFrame([
    # segment, share, purchase_prob, base_items_mean, base_items_sd, addon_prob, price_sensitivity (0-1), category_bias
    {"segment":"Teens/Students",      "share":0.28, "purchase_prob":0.62, "items_mean":1.05, "items_sd":0.25, "addon_prob":0.38, "price_sensitivity":0.70,
     "milkshake_pref":{"classic":0.55,"premium":0.30,"functional":0.05,"dietary":0.10},
     "cake_pref":{"slice":0.55,"traybake":0.25,"snack":0.18,"whole_cake":0.02}},
    {"segment":"Young Professionals", "share":0.26, "purchase_prob":0.66, "items_mean":1.10, "items_sd":0.30, "addon_prob":0.32, "price_sensitivity":0.45,
     "milkshake_pref":{"classic":0.35,"premium":0.40,"functional":0.15,"dietary":0.10},
     "cake_pref":{"slice":0.60,"traybake":0.15,"snack":0.20,"whole_cake":0.05}},
    {"segment":"Families",            "share":0.24, "purchase_prob":0.58, "items_mean":1.35, "items_sd":0.45, "addon_prob":0.44, "price_sensitivity":0.55,
     "milkshake_pref":{"classic":0.50,"premium":0.35,"functional":0.03,"dietary":0.12},
     "cake_pref":{"slice":0.50,"traybake":0.18,"snack":0.22,"whole_cake":0.10}},
    {"segment":"Tourists/Day-out",    "share":0.22, "purchase_prob":0.64, "items_mean":1.15, "items_sd":0.35, "addon_prob":0.40, "price_sensitivity":0.50,
     "milkshake_pref":{"classic":0.40,"premium":0.45,"functional":0.02,"dietary":0.13},
     "cake_pref":{"slice":0.52,"traybake":0.20,"snack":0.18,"whole_cake":0.10}},
])

# Helper: choose menu item based on category preference
def weighted_choice(rng, items, weights):
    w = np.array(weights, dtype=float)
    w = w / w.sum()
    idx = rng.choice(len(items), p=w)
    return items[idx]

def pick_item_for_shop(rng, shop_type, menu_df, seg_row):
    if shop_type == "milkshake":
        prefs = seg_row["milkshake_pref"]
        # Map category -> rows
        cat_to_rows = {
            "classic": menu_df[menu_df["category"]=="classic"],
            "premium": menu_df[menu_df["category"]=="premium"],
            "functional": menu_df[menu_df["category"]=="functional"],
            "dietary": menu_df[menu_df["category"]=="dietary"],
        }
        cat = weighted_choice(rng, list(prefs.keys()), list(prefs.values()))
        pick = cat_to_rows[cat].sample(1, random_state=int(rng.integers(0, 1_000_000))).iloc[0]
        return pick
    else:
        prefs = seg_row["cake_pref"]
        cat_to_rows = {
            "slice": menu_df[menu_df["category"]=="slice"],
            "traybake": menu_df[menu_df["category"]=="traybake"],
            "snack": menu_df[menu_df["category"]=="snack"],
            "whole_cake": menu_df[menu_df["category"]=="whole_cake"],
        }
        cat = weighted_choice(rng, list(prefs.keys()), list(prefs.values()))
        pick = cat_to_rows[cat].sample(1, random_state=int(rng.integers(0, 1_000_000))).iloc[0]
        return pick

def maybe_add_addon(rng, menu_df):
    addons = menu_df[menu_df["category"]=="addon"]
    if len(addons)==0:
        return None
    return addons.sample(1, random_state=int(rng.integers(0, 1_000_000))).iloc[0]


# ----------------------------
# 3) OPERATING ASSUMPTIONS (UK)
# ----------------------------
# You can tune these freely to represent "whole UK" averages or a specific city.
assumptions = {
    "days_per_year": 365,
    "open_days_per_week": 7,
    "base_daily_footfall": 220,   # people passing who could be potential customers
    "weekday_multiplier": 1.00,
    "weekend_multiplier": 1.25,
    "seasonality_strength": 0.10, # +/- 10% across the year (simple sine wave)

    # Labour: use a barista-like hourly reference (Indeed reports ~12.99/hr) :contentReference[oaicite:2]{index=2}
    "hourly_wage": 12.99,
    "hours_open_per_day": 10,
    "staff_on_shift": 2.2,        # average across day (e.g., 2 staff most of day, 3 at peaks)
    "employer_oncost_rate": 0.12, # NI/pension/holiday rough combined factor (adjust as needed)

    # Overheads (rent, utilities, insurance, POS, waste, marketing, etc.)
    "weekly_overheads_gbp": 1850, # tune by location; London/prime sites higher

    # Waste/spoilage allowance (especially relevant for cake)
    "waste_rate_milkshake": 0.03,
    "waste_rate_cake": 0.06,

    # Card fees / payment processing
    "processing_fee_rate": 0.013,
}

# ----------------------------
# 4) SIMULATION ENGINE
# ----------------------------

def simulate_shop(shop_type, menu_df, rng_seed=7, assumptions=assumptions, n_years=1):
    rng = np.random.default_rng(rng_seed)

    days = assumptions["days_per_year"] * n_years
    dates = pd.date_range("2025-01-01", periods=days, freq="D")  # arbitrary start date
    rows = []

    # Pre-split for speed
    segs = segments.copy()
    seg_weights = segs["share"].to_numpy()

    for i, d in enumerate(dates):
        dow = d.dayofweek  # Mon=0 ... Sun=6
        is_weekend = dow >= 5

        # seasonality: simple sine wave across year
        season = 1.0 + assumptions["seasonality_strength"] * np.sin(2*np.pi*(d.dayofyear/365.0))

        day_mult = assumptions["weekend_multiplier"] if is_weekend else assumptions["weekday_multiplier"]
        expected_footfall = assumptions["base_daily_footfall"] * day_mult * season

        # Poisson footfall: potential customers walking by / in the area
        footfall = rng.poisson(lam=max(expected_footfall, 1))

        # segment assignment for footfall
        seg_idx = rng.choice(len(segs), size=footfall, p=seg_weights)
        day_revenue = 0.0
        day_cogs = 0.0
        day_prep_minutes = 0.0
        transactions = 0
        items_sold = 0

        for idx in seg_idx:
            seg = segs.iloc[idx]
            if rng.random() > seg["purchase_prob"]:
                continue

            # number of base items in basket (at least 1)
            n_items = int(max(1, round(rng.normal(seg["items_mean"], seg["items_sd"]))))

            basket_rev = 0.0
            basket_cogs = 0.0
            basket_prep = 0.0

            for _ in range(n_items):
                item = pick_item_for_shop(rng, shop_type, menu_df, seg)
                basket_rev += float(item["price_gbp"])
                basket_cogs += float(item["cogs_gbp"])
                basket_prep += float(item["prep_min"])
                items_sold += 1

                # add-on behaviour
                if rng.random() < seg["addon_prob"]:
                    addon = maybe_add_addon(rng, menu_df)
                    if addon is not None:
                        basket_rev += float(addon["price_gbp"])
                        basket_cogs += float(addon["cogs_gbp"])
                        basket_prep += float(addon["prep_min"])
                        items_sold += 1

            # Payment processing fee
            basket_processing = basket_rev * assumptions["processing_fee_rate"]

            day_revenue += basket_rev
            day_cogs += basket_cogs + basket_processing
            day_prep_minutes += basket_prep
            transactions += 1

        # Waste/spoilage
        waste_rate = assumptions["waste_rate_milkshake"] if shop_type == "milkshake" else assumptions["waste_rate_cake"]
        day_cogs *= (1.0 + waste_rate)

        rows.append({
            "date": d,
            "dow": dow,
            "is_weekend": is_weekend,
            "footfall": footfall,
            "transactions": transactions,
            "items_sold": items_sold,
            "revenue_gbp": day_revenue,
            "cogs_plus_fees_gbp": day_cogs,
            "prep_minutes": day_prep_minutes,
        })

    df = pd.DataFrame(rows)

    # Labour costs: fixed schedule approach (simpler + common for cafes)
    daily_wage = assumptions["hourly_wage"] * assumptions["hours_open_per_day"] * assumptions["staff_on_shift"]
    daily_wage *= (1.0 + assumptions["employer_oncost_rate"])
    df["labour_gbp"] = daily_wage

    # Overheads: weekly overhead spread across days
    df["overheads_gbp"] = assumptions["weekly_overheads_gbp"] / 7.0

    # Profit
    df["gross_profit_gbp"] = df["revenue_gbp"] - df["cogs_plus_fees_gbp"]
    df["net_profit_gbp"] = df["gross_profit_gbp"] - df["labour_gbp"] - df["overheads_gbp"]

    return df


milkshake_df = simulate_shop("milkshake", milkshake_menu, rng_seed=42)
cake_df = simulate_shop("cake", cake_menu, rng_seed=43)

# ----------------------------
# 5) SUMMARY: weekly + yearly
# ----------------------------

def summarize(df, label):
    weekly = df.resample("W", on="date")[["revenue_gbp","cogs_plus_fees_gbp","labour_gbp","overheads_gbp","net_profit_gbp"]].sum()
    yearly = df.resample("Y", on="date")[["revenue_gbp","cogs_plus_fees_gbp","labour_gbp","overheads_gbp","net_profit_gbp"]].sum()

    out = {
        "business": label,
        "avg_weekly_revenue": weekly["revenue_gbp"].mean(),
        "avg_weekly_net_profit": weekly["net_profit_gbp"].mean(),
        "year_revenue": float(yearly["revenue_gbp"].iloc[0]),
        "year_net_profit": float(yearly["net_profit_gbp"].iloc[0]),
        "avg_gross_margin": float((df["gross_profit_gbp"].sum() / max(df["revenue_gbp"].sum(), 1)) ),
        "avg_net_margin": float((df["net_profit_gbp"].sum() / max(df["revenue_gbp"].sum(), 1)) ),
        "avg_daily_transactions": df["transactions"].mean(),
        "avg_ticket_size": float(df["revenue_gbp"].sum() / max(df["transactions"].sum(), 1)),
    }
    return weekly, yearly, out

milk_weekly, milk_yearly, milk_kpis = summarize(milkshake_df, "Milkshake Shop")
cake_weekly, cake_yearly, cake_kpis = summarize(cake_df, "Cake Shop")

kpis = pd.DataFrame([milk_kpis, cake_kpis])
display(kpis)

# ----------------------------
# 6) SIDE-BY-SIDE COMPARISON CHARTS
# ----------------------------

def plot_weekly(weekly_a, weekly_b, label_a="Milkshake", label_b="Cake"):
    plt.figure()
    plt.plot(weekly_a.index, weekly_a["net_profit_gbp"], label=label_a)
    plt.plot(weekly_b.index, weekly_b["net_profit_gbp"], label=label_b)
    plt.title("Weekly Net Profit (simulated)")
    plt.xlabel("Week")
    plt.ylabel("Net profit ()")
    plt.legend()
    plt.show()

    plt.figure()
    plt.plot(weekly_a.index, weekly_a["revenue_gbp"], label=label_a)
    plt.plot(weekly_b.index, weekly_b["revenue_gbp"], label=label_b)
    plt.title("Weekly Revenue (simulated)")
    plt.xlabel("Week")
    plt.ylabel("Revenue ()")
    plt.legend()
    plt.show()

plot_weekly(milk_weekly, cake_weekly)

# ----------------------------
# 7) MENU ECONOMICS: item margins & prep time pressure
# ----------------------------

def menu_economics(menu_df, name):
    m = menu_df.copy()
    m["gross_profit_gbp"] = m["price_gbp"] - m["cogs_gbp"]
    m["gross_margin"] = m["gross_profit_gbp"] / m["price_gbp"]
    m["gp_per_min"] = m["gross_profit_gbp"] / m["prep_min"]
    m = m.sort_values(["gross_margin","gp_per_min"], ascending=False)
    print(f"\n=== {name} menu economics (top rows) ===")
    display(m.head(10))
    return m

milk_menu_econ = menu_economics(milkshake_menu, "Milkshake Shop")
cake_menu_econ = menu_economics(cake_menu, "Cake Shop")

# ----------------------------
# 8) EXPLAINABLE "LOGIC" COMPARISON (quick table)
# ----------------------------

logic_compare = pd.DataFrame([
    {"dimension":"Core product", "Milkshake Shop":"Cold blended drinks + add-ons", "Cake Shop":"Slices, traybakes, snacks, celebration cakes"},
    {"dimension":"COGS profile", "Milkshake Shop":"Lower spoilage, dairy/syrup driven", "Cake Shop":"Higher spoilage + batch baking ingredients"},
    {"dimension":"Peak demand", "Milkshake Shop":"Afternoons/evenings; warm days", "Cake Shop":"All day; strong coffee-time pairing"},
    {"dimension":"Speed / throughput", "Milkshake Shop":"Prep minutes can bottleneck at peaks", "Cake Shop":"Fast serve (display cabinet)"},
    {"dimension":"Typical upsell", "Milkshake Shop":"Toppings, sizes, premium mix-ins", "Cake Shop":"Ice cream/sauces, bigger slices, whole cakes"},
    {"dimension":"Demographics pull", "Milkshake Shop":"Younger + treat occasions", "Cake Shop":"Broader; families + office visitors"},
    {"dimension":"Margin strategy target", "Milkshake Shop":"Aim for ~70% GP overall", "Cake Shop":"Aim for ~70% GP overall"},
])

display(logic_compare)

# ----------------------------
# 9) WHAT-IF TOOLS
# ----------------------------

def run_what_if(base_daily_footfall=None, weekly_overheads_gbp=None, hourly_wage=None):
    new_assumptions = assumptions.copy()
    if base_daily_footfall is not None:
        new_assumptions["base_daily_footfall"] = base_daily_footfall
    if weekly_overheads_gbp is not None:
        new_assumptions["weekly_overheads_gbp"] = weekly_overheads_gbp
    if hourly_wage is not None:
        new_assumptions["hourly_wage"] = hourly_wage

    milk = simulate_shop("milkshake", milkshake_menu, rng_seed=101, assumptions=new_assumptions)
    cake = simulate_shop("cake", cake_menu, rng_seed=102, assumptions=new_assumptions)

    _, _, mk = summarize(milk, "Milkshake Shop")
    _, _, ck = summarize(cake, "Cake Shop")

    return pd.DataFrame([mk, ck])

# Example: higher rent + higher footfall (e.g., prime shopping centre)
display(run_what_if(base_daily_footfall=320, weekly_overheads_gbp=2800, hourly_wage=12.99))
```

```python
# JUPYTER NOTEBOOK CELL: Milkshake vs Cake Shop (UK)  menus, demographics, and profit simulation

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------------
# 1) MENUS (synthetic but realistic-leaning)
# ----------------------------

def build_menus():
    milkshake_menu = pd.DataFrame([
        {"item":"Classic Vanilla Shake (12oz)", "category":"classic", "price_gbp":4.95, "cogs_gbp":1.25, "prep_min":3},
        {"item":"Classic Chocolate Shake (12oz)", "category":"classic", "price_gbp":4.95, "cogs_gbp":1.30, "prep_min":3},
        {"item":"Strawberry Shake (12oz)", "category":"classic", "price_gbp":5.25, "cogs_gbp":1.35, "prep_min":3},
        {"item":"Salted Caramel Shake (16oz)", "category":"premium", "price_gbp":6.25, "cogs_gbp":1.85, "prep_min":4},
        {"item":"Oreo / Cookies Shake (16oz)", "category":"premium", "price_gbp":6.75, "cogs_gbp":2.10, "prep_min":4},
        {"item":"Protein Shake (16oz)", "category":"functional", "price_gbp":7.25, "cogs_gbp":2.55, "prep_min":4},
        {"item":"Vegan Shake (oat-based, 16oz)", "category":"dietary", "price_gbp":6.95, "cogs_gbp":2.35, "prep_min":4},
        {"item":"Add-on: Whipped cream", "category":"addon", "price_gbp":0.75, "cogs_gbp":0.18, "prep_min":0.5},
        {"item":"Add-on: Extra syrup", "category":"addon", "price_gbp":0.60, "cogs_gbp":0.12, "prep_min":0.2},
        {"item":"Add-on: Extra topping (sprinkles/cookie)", "category":"addon", "price_gbp":0.85, "cogs_gbp":0.25, "prep_min":0.3},
    ])

    cake_menu = pd.DataFrame([
        {"item":"Victoria Sponge (slice)", "category":"slice", "price_gbp":3.95, "cogs_gbp":1.10, "prep_min":1.2},
        {"item":"Chocolate Fudge Cake (slice)", "category":"slice", "price_gbp":4.25, "cogs_gbp":1.25, "prep_min":1.2},
        {"item":"Carrot Cake (slice)", "category":"slice", "price_gbp":4.35, "cogs_gbp":1.30, "prep_min":1.2},
        {"item":"Cheesecake (slice)", "category":"slice", "price_gbp":4.65, "cogs_gbp":1.55, "prep_min":1.2},
        {"item":"Brownie", "category":"traybake", "price_gbp":3.25, "cogs_gbp":0.85, "prep_min":1.0},
        {"item":"Cookie", "category":"snack", "price_gbp":2.25, "cogs_gbp":0.55, "prep_min":0.8},
        {"item":"Cupcake", "category":"snack", "price_gbp":2.95, "cogs_gbp":0.70, "prep_min":1.0},
        {"item":"Celebration Cake (whole, 10 slices)", "category":"whole_cake", "price_gbp":34.00, "cogs_gbp":11.50, "prep_min":6.0},
        {"item":"Add-on: Ice cream scoop", "category":"addon", "price_gbp":1.50, "cogs_gbp":0.40, "prep_min":0.3},
        {"item":"Add-on: Sauce drizzle", "category":"addon", "price_gbp":0.70, "cogs_gbp":0.15, "prep_min":0.2},
    ])

    return milkshake_menu, cake_menu


milkshake_menu, cake_menu = build_menus()

display(milkshake_menu)
display(cake_menu)

# ----------------------------
# 2) CUSTOMER DEMOGRAPHICS + BUYING LOGIC
# ----------------------------
# We'll model the UK customer mix as segments with:
# - share of footfall
# - probability of purchase
# - typical basket size + add-on behaviour
# - preferred categories

segments = pd.DataFrame([
    # segment, share, purchase_prob, base_items_mean, base_items_sd, addon_prob, price_sensitivity (0-1), category_bias
    {"segment":"Teens/Students",      "share":0.28, "purchase_prob":0.62, "items_mean":1.05, "items_sd":0.25, "addon_prob":0.38, "price_sensitivity":0.70,
     "milkshake_pref":{"classic":0.55,"premium":0.30,"functional":0.05,"dietary":0.10},
     "cake_pref":{"slice":0.55,"traybake":0.25,"snack":0.18,"whole_cake":0.02}},
    {"segment":"Young Professionals", "share":0.26, "purchase_prob":0.66, "items_mean":1.10, "items_sd":0.30, "addon_prob":0.32, "price_sensitivity":0.45,
     "milkshake_pref":{"classic":0.35,"premium":0.40,"functional":0.15,"dietary":0.10},
     "cake_pref":{"slice":0.60,"traybake":0.15,"snack":0.20,"whole_cake":0.05}},
    {"segment":"Families",            "share":0.24, "purchase_prob":0.58, "items_mean":1.35, "items_sd":0.45, "addon_prob":0.44, "price_sensitivity":0.55,
     "milkshake_pref":{"classic":0.50,"premium":0.35,"functional":0.03,"dietary":0.12},
     "cake_pref":{"slice":0.50,"traybake":0.18,"snack":0.22,"whole_cake":0.10}},
    {"segment":"Tourists/Day-out",    "share":0.22, "purchase_prob":0.64, "items_mean":1.15, "items_sd":0.35, "addon_prob":0.40, "price_sensitivity":0.50,
     "milkshake_pref":{"classic":0.40,"premium":0.45,"functional":0.02,"dietary":0.13},
     "cake_pref":{"slice":0.52,"traybake":0.20,"snack":0.18,"whole_cake":0.10}},
])

# Helper: choose menu item based on category preference
def weighted_choice(rng, items, weights):
    w = np.array(weights, dtype=float)
    w = w / w.sum()
    idx = rng.choice(len(items), p=w)
    return items[idx]

def pick_item_for_shop(rng, shop_type, menu_df, seg_row):
    if shop_type == "milkshake":
        prefs = seg_row["milkshake_pref"]
        # Map category -> rows
        cat_to_rows = {
            "classic": menu_df[menu_df["category"]=="classic"],
            "premium": menu_df[menu_df["category"]=="premium"],
            "functional": menu_df[menu_df["category"]=="functional"],
            "dietary": menu_df[menu_df["category"]=="dietary"],
        }
        cat = weighted_choice(rng, list(prefs.keys()), list(prefs.values()))
        pick = cat_to_rows[cat].sample(1, random_state=int(rng.integers(0, 1_000_000))).iloc[0]
        return pick
    else:
        prefs = seg_row["cake_pref"]
        cat_to_rows = {
            "slice": menu_df[menu_df["category"]=="slice"],
            "traybake": menu_df[menu_df["category"]=="traybake"],
            "snack": menu_df[menu_df["category"]=="snack"],
            "whole_cake": menu_df[menu_df["category"]=="whole_cake"],
        }
        cat = weighted_choice(rng, list(prefs.keys()), list(prefs.values()))
        pick = cat_to_rows[cat].sample(1, random_state=int(rng.integers(0, 1_000_000))).iloc[0]
        return pick

def maybe_add_addon(rng, menu_df):
    addons = menu_df[menu_df["category"]=="addon"]
    if len(addons)==0:
        return None
    return addons.sample(1, random_state=int(rng.integers(0, 1_000_000))).iloc[0]


# ----------------------------
# 3) OPERATING ASSUMPTIONS (UK)
# ----------------------------
# You can tune these freely to represent "whole UK" averages or a specific city.
assumptions = {
    "days_per_year": 365,
    "open_days_per_week": 7,
    "base_daily_footfall": 220,   # people passing who could be potential customers
    "weekday_multiplier": 1.00,
    "weekend_multiplier": 1.25,
    "seasonality_strength": 0.10, # +/- 10% across the year (simple sine wave)

    # Labour: use a barista-like hourly reference (Indeed reports ~12.99/hr) :contentReference[oaicite:2]{index=2}
    "hourly_wage": 12.99,
    "hours_open_per_day": 10,
    "staff_on_shift": 2.2,        # average across day (e.g., 2 staff most of day, 3 at peaks)
    "employer_oncost_rate": 0.12, # NI/pension/holiday rough combined factor (adjust as needed)

    # Overheads (rent, utilities, insurance, POS, waste, marketing, etc.)
    "weekly_overheads_gbp": 1850, # tune by location; London/prime sites higher

    # Waste/spoilage allowance (especially relevant for cake)
    "waste_rate_milkshake": 0.03,
    "waste_rate_cake": 0.06,

    # Card fees / payment processing
    "processing_fee_rate": 0.013,
}

# ----------------------------
# 4) SIMULATION ENGINE
# ----------------------------

def simulate_shop(shop_type, menu_df, rng_seed=7, assumptions=assumptions, n_years=1):
    rng = np.random.default_rng(rng_seed)

    days = assumptions["days_per_year"] * n_years
    dates = pd.date_range("2025-01-01", periods=days, freq="D")  # arbitrary start date
    rows = []

    # Pre-split for speed
    segs = segments.copy()
    seg_weights = segs["share"].to_numpy()

    for i, d in enumerate(dates):
        dow = d.dayofweek  # Mon=0 ... Sun=6
        is_weekend = dow >= 5

        # seasonality: simple sine wave across year
        season = 1.0 + assumptions["seasonality_strength"] * np.sin(2*np.pi*(d.dayofyear/365.0))

        day_mult = assumptions["weekend_multiplier"] if is_weekend else assumptions["weekday_multiplier"]
        expected_footfall = assumptions["base_daily_footfall"] * day_mult * season

        # Poisson footfall: potential customers walking by / in the area
        footfall = rng.poisson(lam=max(expected_footfall, 1))

        # segment assignment for footfall
        seg_idx = rng.choice(len(segs), size=footfall, p=seg_weights)
        day_revenue = 0.0
        day_cogs = 0.0
        day_prep_minutes = 0.0
        transactions = 0
        items_sold = 0

        for idx in seg_idx:
            seg = segs.iloc[idx]
            if rng.random() > seg["purchase_prob"]:
                continue

            # number of base items in basket (at least 1)
            n_items = int(max(1, round(rng.normal(seg["items_mean"], seg["items_sd"]))))

            basket_rev = 0.0
            basket_cogs = 0.0
            basket_prep = 0.0

            for _ in range(n_items):
                item = pick_item_for_shop(rng, shop_type, menu_df, seg)
                basket_rev += float(item["price_gbp"])
                basket_cogs += float(item["cogs_gbp"])
                basket_prep += float(item["prep_min"])
                items_sold += 1

                # add-on behaviour
                if rng.random() < seg["addon_prob"]:
                    addon = maybe_add_addon(rng, menu_df)
                    if addon is not None:
                        basket_rev += float(addon["price_gbp"])
                        basket_cogs += float(addon["cogs_gbp"])
                        basket_prep += float(addon["prep_min"])
                        items_sold += 1

            # Payment processing fee
            basket_processing = basket_rev * assumptions["processing_fee_rate"]

            day_revenue += basket_rev
            day_cogs += basket_cogs + basket_processing
            day_prep_minutes += basket_prep
            transactions += 1

        # Waste/spoilage
        waste_rate = assumptions["waste_rate_milkshake"] if shop_type == "milkshake" else assumptions["waste_rate_cake"]
        day_cogs *= (1.0 + waste_rate)

        rows.append({
            "date": d,
            "dow": dow,
            "is_weekend": is_weekend,
            "footfall": footfall,
            "transactions": transactions,
            "items_sold": items_sold,
            "revenue_gbp": day_revenue,
            "cogs_plus_fees_gbp": day_cogs,
            "prep_minutes": day_prep_minutes,
        })

    df = pd.DataFrame(rows)

    # Labour costs: fixed schedule approach (simpler + common for cafes)
    daily_wage = assumptions["hourly_wage"] * assumptions["hours_open_per_day"] * assumptions["staff_on_shift"]
    daily_wage *= (1.0 + assumptions["employer_oncost_rate"])
    df["labour_gbp"] = daily_wage

    # Overheads: weekly overhead spread across days
    df["overheads_gbp"] = assumptions["weekly_overheads_gbp"] / 7.0

    # Profit
    df["gross_profit_gbp"] = df["revenue_gbp"] - df["cogs_plus_fees_gbp"]
    df["net_profit_gbp"] = df["gross_profit_gbp"] - df["labour_gbp"] - df["overheads_gbp"]

    return df


milkshake_df = simulate_shop("milkshake", milkshake_menu, rng_seed=42)
cake_df = simulate_shop("cake", cake_menu, rng_seed=43)

# ----------------------------
# 5) SUMMARY: weekly + yearly
# ----------------------------

def summarize(df, label):
    weekly = df.resample("W", on="date")[["revenue_gbp","cogs_plus_fees_gbp","labour_gbp","overheads_gbp","net_profit_gbp"]].sum()
    yearly = df.resample("Y", on="date")[["revenue_gbp","cogs_plus_fees_gbp","labour_gbp","overheads_gbp","net_profit_gbp"]].sum()

    out = {
        "business": label,
        "avg_weekly_revenue": weekly["revenue_gbp"].mean(),
        "avg_weekly_net_profit": weekly["net_profit_gbp"].mean(),
        "year_revenue": float(yearly["revenue_gbp"].iloc[0]),
        "year_net_profit": float(yearly["net_profit_gbp"].iloc[0]),
        "avg_gross_margin": float((df["gross_profit_gbp"].sum() / max(df["revenue_gbp"].sum(), 1)) ),
        "avg_net_margin": float((df["net_profit_gbp"].sum() / max(df["revenue_gbp"].sum(), 1)) ),
        "avg_daily_transactions": df["transactions"].mean(),
        "avg_ticket_size": float(df["revenue_gbp"].sum() / max(df["transactions"].sum(), 1)),
    }
    return weekly, yearly, out

milk_weekly, milk_yearly, milk_kpis = summarize(milkshake_df, "Milkshake Shop")
cake_weekly, cake_yearly, cake_kpis = summarize(cake_df, "Cake Shop")

kpis = pd.DataFrame([milk_kpis, cake_kpis])
display(kpis)

# ----------------------------
# 6) SIDE-BY-SIDE COMPARISON CHARTS
# ----------------------------

def plot_weekly(weekly_a, weekly_b, label_a="Milkshake", label_b="Cake"):
    plt.figure()
    plt.plot(weekly_a.index, weekly_a["net_profit_gbp"], label=label_a)
    plt.plot(weekly_b.index, weekly_b["net_profit_gbp"], label=label_b)
    plt.title("Weekly Net Profit (simulated)")
    plt.xlabel("Week")
    plt.ylabel("Net profit ()")
    plt.legend()
    plt.show()

    plt.figure()
    plt.plot(weekly_a.index, weekly_a["revenue_gbp"], label=label_a)
    plt.plot(weekly_b.index, weekly_b["revenue_gbp"], label=label_b)
    plt.title("Weekly Revenue (simulated)")
    plt.xlabel("Week")
    plt.ylabel("Revenue ()")
    plt.legend()
    plt.show()

plot_weekly(milk_weekly, cake_weekly)

# ----------------------------
# 7) MENU ECONOMICS: item margins & prep time pressure
# ----------------------------

def menu_economics(menu_df, name):
    m = menu_df.copy()
    m["gross_profit_gbp"] = m["price_gbp"] - m["cogs_gbp"]
    m["gross_margin"] = m["gross_profit_gbp"] / m["price_gbp"]
    m["gp_per_min"] = m["gross_profit_gbp"] / m["prep_min"]
    m = m.sort_values(["gross_margin","gp_per_min"], ascending=False)
    print(f"\n=== {name} menu economics (top rows) ===")
    display(m.head(10))
    return m

milk_menu_econ = menu_economics(milkshake_menu, "Milkshake Shop")
cake_menu_econ = menu_economics(cake_menu, "Cake Shop")

# ----------------------------
# 8) EXPLAINABLE "LOGIC" COMPARISON (quick table)
# ----------------------------

logic_compare = pd.DataFrame([
    {"dimension":"Core product", "Milkshake Shop":"Cold blended drinks + add-ons", "Cake Shop":"Slices, traybakes, snacks, celebration cakes"},
    {"dimension":"COGS profile", "Milkshake Shop":"Lower spoilage, dairy/syrup driven", "Cake Shop":"Higher spoilage + batch baking ingredients"},
    {"dimension":"Peak demand", "Milkshake Shop":"Afternoons/evenings; warm days", "Cake Shop":"All day; strong coffee-time pairing"},
    {"dimension":"Speed / throughput", "Milkshake Shop":"Prep minutes can bottleneck at peaks", "Cake Shop":"Fast serve (display cabinet)"},
    {"dimension":"Typical upsell", "Milkshake Shop":"Toppings, sizes, premium mix-ins", "Cake Shop":"Ice cream/sauces, bigger slices, whole cakes"},
    {"dimension":"Demographics pull", "Milkshake Shop":"Younger + treat occasions", "Cake Shop":"Broader; families + office visitors"},
    {"dimension":"Margin strategy target", "Milkshake Shop":"Aim for ~70% GP overall", "Cake Shop":"Aim for ~70% GP overall"},
])

display(logic_compare)

# ----------------------------
# 9) WHAT-IF TOOLS
# ----------------------------

def run_what_if(base_daily_footfall=None, weekly_overheads_gbp=None, hourly_wage=None):
    new_assumptions = assumptions.copy()
    if base_daily_footfall is not None:
        new_assumptions["base_daily_footfall"] = base_daily_footfall
    if weekly_overheads_gbp is not None:
        new_assumptions["weekly_overheads_gbp"] = weekly_overheads_gbp
    if hourly_wage is not None:
        new_assumptions["hourly_wage"] = hourly_wage

    milk = simulate_shop("milkshake", milkshake_menu, rng_seed=101, assumptions=new_assumptions)
    cake = simulate_shop("cake", cake_menu, rng_seed=102, assumptions=new_assumptions)

    _, _, mk = summarize(milk, "Milkshake Shop")
    _, _, ck = summarize(cake, "Cake Shop")

    return pd.DataFrame([mk, ck])

# Example: higher rent + higher footfall (e.g., prime shopping centre)
display(run_what_if(base_daily_footfall=320, weekly_overheads_gbp=2800, hourly_wage=12.99))
```

```python
# ----------------------------
# REGIONAL ASSUMPTIONS (UK)
# ----------------------------

regions = {
    "England": {
        "footfall_multiplier": 1.00,
        "weekly_overheads_gbp": 1900,
        "hourly_wage": 13.20
    },
    "Scotland": {
        "footfall_multiplier": 0.85,
        "weekly_overheads_gbp": 1500,
        "hourly_wage": 12.40
    },
    "Wales": {
        "footfall_multiplier": 0.80,
        "weekly_overheads_gbp": 1350,
        "hourly_wage": 12.20
    },
    "Northern Ireland": {
        "footfall_multiplier": 0.75,
        "weekly_overheads_gbp": 1250,
        "hourly_wage": 11.90
    }
}

def run_region_simulation(shop_type, menu_df):
    outputs = []

    for region, params in regions.items():
        region_assumptions = assumptions.copy()
        region_assumptions["base_daily_footfall"] *= params["footfall_multiplier"]
        region_assumptions["weekly_overheads_gbp"] = params["weekly_overheads_gbp"]
        region_assumptions["hourly_wage"] = params["hourly_wage"]

        df = simulate_shop(
            shop_type,
            menu_df,
            rng_seed=hash(region) % 10_000,
            assumptions=region_assumptions
        )

        _, _, kpi = summarize(df, f"{shop_type.title()}  {region}")
        kpi["region"] = region
        outputs.append(kpi)

    return pd.DataFrame(outputs)

milkshake_regions = run_region_simulation("milkshake", milkshake_menu)
cake_regions = run_region_simulation("cake", cake_menu)

display(milkshake_regions)
display(cake_regions)
```

```python
# ----------------------------
# DELIVERY CHANNELS
# ----------------------------

delivery_channels = {
    "Uber Eats": {
        "order_share": 0.22,
        "commission_rate": 0.30,
        "packaging_cost_gbp": 0.35
    },
    "Deliveroo": {
        "order_share": 0.18,
        "commission_rate": 0.28,
        "packaging_cost_gbp": 0.30
    }
}

def apply_delivery_channel(df, channel_name):
    channel = delivery_channels[channel_name]
    df = df.copy()

    delivery_orders = df["transactions"] * channel["order_share"]

    delivery_revenue = df["revenue_gbp"] * channel["order_share"]
    commission_cost = delivery_revenue * channel["commission_rate"]
    packaging_cost = delivery_orders * channel["packaging_cost_gbp"]

    df["delivery_revenue_gbp"] = delivery_revenue
    df["delivery_costs_gbp"] = commission_cost + packaging_cost

    df["net_profit_gbp"] -= df["delivery_costs_gbp"]

    return df

# Apply both platforms
milkshake_delivery = apply_delivery_channel(
    apply_delivery_channel(milkshake_df, "Uber Eats"),
    "Deliveroo"
)

cake_delivery = apply_delivery_channel(
    apply_delivery_channel(cake_df, "Uber Eats"),
    "Deliveroo"
)

_, _, milk_delivery_kpi = summarize(milkshake_delivery, "Milkshake + Delivery")
_, _, cake_delivery_kpi = summarize(cake_delivery, "Cake + Delivery")

display(pd.DataFrame([milk_delivery_kpi, cake_delivery_kpi]))
```

```python
# ----------------------------
# VAT VIEW (UK)
# ----------------------------

VAT_RATE = 0.20

def apply_vat_view(df):
    df = df.copy()

    df["revenue_net_vat"] = df["revenue_gbp"] / (1 + VAT_RATE)
    df["vat_payable"] = df["revenue_gbp"] - df["revenue_net_vat"]

    df["gross_profit_net_vat"] = (
        df["revenue_net_vat"] - df["cogs_plus_fees_gbp"]
    )

    df["net_profit_net_vat"] = (
        df["gross_profit_net_vat"]
        - df["labour_gbp"]
        - df["overheads_gbp"]
    )

    return df

milkshake_vat = apply_vat_view(milkshake_delivery)
cake_vat = apply_vat_view(cake_delivery)

def vat_summary(df, label):
    return {
        "business": label,
        "year_revenue_gross": df["revenue_gbp"].sum(),
        "year_revenue_net_vat": df["revenue_net_vat"].sum(),
        "vat_paid_to_hmrc": df["vat_payable"].sum(),
        "year_net_profit_net_vat": df["net_profit_net_vat"].sum(),
        "net_margin_net_vat": df["net_profit_net_vat"].sum() / df["revenue_net_vat"].sum()
    }

vat_kpis = pd.DataFrame([
    vat_summary(milkshake_vat, "Milkshake Shop"),
    vat_summary(cake_vat, "Cake Shop")
])

display(vat_kpis)
```
